```python
# max 함수
print(data.loc[data['광역시도'] == '강원도', 'population'].max())
print(data.loc[data['광역시도'] == '강원도', 'population'].min())
print(data.loc[data['광역시도'] == '강원도', 'population'].mean())
```

```python
max_pop = data.loc[data['광역시도'] == '강원도', 'population'].max()
max_pop
```

```python
data.loc[data['population'] == max_pop, ['광역시도', '행정구역', 'population']]
```

```python
def get_max_pop_info(sido_name):
    max_pop = data.loc[data['광역시도'] == sido_name, 'population'].max()
    return data.loc[data['population'] == max_pop, ['광역시도', '행정구역', 'population']]
```

```python
get_max_pop_info('부산광역시')
```

```python
#empty DataFrame 생성
max_pop_df = pd.DataFrame(columns=['광역시도', '행정구역', 'population'])
max_pop_df
```

```python
for sido_name in data['광역시도'].unique():
    df = get_max_pop_info(sido_name)
    max_pop_df = max_pop_df.append(df)

max_pop_df
```

```python
get_max_pop_info('경상북도')
```

```python
#reset index
max_pop_df = max_pop_df.reset_index()
```

```python
max_pop_df
```

```python
# delete column : drop()
max_pop_df.drop('index', axis=1, inplace=True)
```

```python
max_pop_df
```

```python
# DataFrame을 excel file로 저장
max_pop_df.to_excel('data/maxpop.xlsx', sheet_name='광역시도별 최대인구')
```
